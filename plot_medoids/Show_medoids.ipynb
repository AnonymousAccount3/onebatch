```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import kmedoids as km
from sklearn.metrics import pairwise_distances
from sklearn.manifold import TSNE

from onebatch import one_batch_greedy_kmedoids
from data_loaders import load_data


matplotlib.rc('axes',linewidth=2)
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
matplotlib.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
matplotlib.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

## MNIST

### K = 10

```python
dataset = "mnist"
N = 20000
K = 10
seed = 0

np.random.seed(seed)
X, _ = load_data(dataset, seed, N)
```

```python
np.random.seed(seed)
indexes = one_batch_greedy_kmedoids(X, K=K,  batch_size=1000, verbose=0)

fig, axes = plt.subplots(2, 5, figsize=(8, 3))
for ax, i in zip(axes.ravel(), indexes):
    ax.imshow(X[i].reshape(28, 28), cmap="gray")
    ax.axis("off")
plt.suptitle("OneBatch - K = %i"%K, fontsize=14)
plt.savefig("medoids/onebatch_mnist_K%i.png"%K, format="png", bbox_inches="tight")
plt.show()
```

```python
np.random.seed(seed)
Dist = pairwise_distances(X, metric="euclidean")

fasterpam = km.KMedoids(n_clusters=K, metric="precomputed", method="fasterpam")
fasterpam.fit(Dist)
indexes = fasterpam.medoid_indices_

fig, axes = plt.subplots(2, 5, figsize=(8, 3))
for ax, i in zip(axes.ravel(), indexes):
    ax.imshow(X[i].reshape(28, 28), cmap="gray")
    ax.axis("off")
plt.suptitle("FasterPAM - K = %i"%K, fontsize=14)
plt.savefig("medoids/fasterpam_mnist_K%i.png"%K, format="png", bbox_inches="tight")
plt.show()
```
